```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
%matplotlib inline


warnings.filterwarnings('ignore')
```

```python
## Read the data
df = pd.read_csv('Bank_Personal_Loan_Modelling.csv')
df = df.drop(['ID', 'ZIP Code'], axis = 1)
df
```

```python
df_test = df.drop(['Personal Loan'], axis = 1)
df_sample = df_test.sample(frac=0.1, replace=True, random_state=1)
df_sample
```

```python
from sklearn.ensemble import RandomForestClassifier
modelRF= RandomForestClassifier(n_estimators=100)
modelRF.fit(df_test,df['Personal Loan'])
```

```python
df_sample['Personal Loan']=modelRF.predict(df_sample)
```

```python
df_sample.rename(columns={"Personal Loan": "Personal Loan_RF"}, inplace = True)
```

```python
from sklearn.linear_model import LogisticRegression

## LOGISTIC REGRESSION DEFAULT
logreg_model = LogisticRegression()
logreg_model.fit(df_test,df['Personal Loan'])
```

```python
df_sample_lr = df_test.sample(frac=0.1, replace=True, random_state=1)
```

```python
df_sample_lr['Personal Loan_LOGREG']=logreg_model.predict(df_sample_lr)
```
